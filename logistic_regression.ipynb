```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('wine_equality.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
#sklearn mannual implementation of the code
#first we peform collinearity for all the columns and remove one of the columns if the collineraity is greater than 0.5
# Calculate correlation matrix
correlation_matrix = data.corr()
correlation_matrix.to_csv('correlation.csv')
```

```python
correlation_matrix
```

<h3><div>removing one of the columns having the correlation less than 0.5</div></h3>

```python
highly_correlated = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.5:
            colname_i = correlation_matrix.columns[i]
            colname_j = correlation_matrix.columns[j]
            highly_correlated.add((colname_i, colname_j))
columns_to_drop = set()
for col1, col2 in highly_correlated:
    columns_to_drop.add(col2)

data_after_removal = data.drop(columns=columns_to_drop, axis=1)
print("the columns which we are dropping are :\n\t",columns_to_drop,"\n")
print("the remaining columns are :\n\t",data_after_removal.columns)
```

<h3><div>the eliminated columns are Citric_acid , Fixed_acidity , Free_sulfur_dioxide , Volatile_acidity</div></h3>

```python
dar = data_after_removal
dar.head()
```

```python
dar.isnull().sum()
```

<h3><div>there is no preprocessing required as the data does not have any null values</div></h3>

```python
dar.head()
```

<h3>first let me use logisitic regression with threshold of quality greater than 6 as good and less than 6 as bad<br></h3>

```python
from sklearn.model_selection import train_test_split
```

```python
X = dar.iloc[:,0:-1]
X
```

```python
dar['quality'].value_counts()
```

```python
dar['quality'] = dar['quality'].map({3 : 'bad', 4 :'bad', 5: 'bad',
                                      6: 'good', 7: 'good', 8: 'good'}) 
dar['quality']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

dar['quality'] = le.fit_transform(dar['quality'])

dar['quality'].value_counts()

y = dar['quality']
y = np.array(y)
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
def logistic_regression(X, y, lr=0.01, epochs=1000):
    m, n = X.shape
    theta = np.zeros(n) 

    for epoch in range(epochs):
        z = np.dot(X, theta)
        h = sigmoid(z)
        error = h - y

        gradient = np.dot(X.T, error) / m
        theta -= lr * gradient

    return theta

theta = logistic_regression(x_train, y_train)
print(theta)
```

```python
count = 0
predicted = []
for i in range(x_test.shape[0]):
    predicted.append(np.dot(x_test.iloc[i],theta))
for i in range(x_test.shape[0]):
    if predicted[i] > 0.6:
        predicted[i] = 1
    else:
        predicted[i] = 0
    if y_test[i] - predicted[i] == 0 :
        count = count + 1
accuracy = (count/x_test.shape[0])*100
print("testing_accuracy : ", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predicted))
```

<h3>we got an accuracy of 58.75% in logistic regression<br></h3>
